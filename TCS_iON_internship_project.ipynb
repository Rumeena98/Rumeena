**Creating Name column with faker**

```python
def create_names(n):
   name = []
   for _ in range(0, n):
        name.append(faker.name())
   return name
```

```python
df_name = pd.DataFrame(list(zip(name)),columns = ['Name'])
```

**Creating Race Column**

```python
def weighted_cartesian_choice(*iterables):
    # An arbitrary number of tuple or lists,each consisting of population and weights.weighted_cartesian_choice returns a list 
    #with a choice from each population
    
    res = []
    for population, weights in iterables:
     #normalize weight:
        weights = np.array(weights) / sum(weights)
        lst = np.random.choice(population, p=weights)
        res.append(lst)
    return res
race = (['White','Hispanic', 'Black','Asian'], 
        [.32,.28,.22,.18])

list=[]
for i in range(362806):
    res = weighted_cartesian_choice(race)
    list.append(res)
```

```python
df_race=pd.DataFrame(list, columns = ['Race'])
```

```python
df_race.head()
```

```python
drug_counts = df['Drug'].value_counts().nlargest(10)
top_drugs_df = df[df['Drug'].isin(drug_counts.index)]
drug_side_counts = top_drugs_df.groupby(['Drug', 'Sides']).size().unstack()
drug_side_counts.plot(kind='bar', stacked=True, figsize=(10, 6))

plt.title('Top 10 Drugs and their Side Effects')
plt.xlabel('Drug')
plt.ylabel('Count')

plt.legend(title='Side Effects', loc='upper right')
plt.xticks(rotation=45)

plt.show()
```

**Inferences:**

From the figure, Cymbalta is the most used drug among people. 

And the most common side effects are Nausea, dry mouth,dizziness,drowsiness and sweating. 

The plot analyses that the side effect increases with the usage of drug. 

```python
drug_counts = df['Drug'].value_counts().nlargest(10)
drug_sex_counts = df.groupby(['Drug', 'Sex']).size().unstack().loc[drug_counts.index]
drug_sex_counts.plot(kind='bar', stacked=True, figsize=(8, 6))
plt.title('Top 10 Drugs Usage by Sex')
plt.xlabel('Drug')
plt.ylabel('Count')
plt.legend(title='Sex', loc='upper right')
plt.xticks(rotation=45)
plt.show()
```

**Inferences:** 

The plot shows the usage of drugs among people.  

Females are on the top in the usage of drugs than males. 

Cymbalta is the most used drug. 

```python
drug_counts = df['Drug'].value_counts().nsmallest(10)
drug_sex_counts = df.groupby(['Drug', 'Sex']).size().unstack().loc[drug_counts.index]
drug_sex_counts.plot(kind='bar', stacked=True, figsize=(8, 6))
plt.title('Least Drugs Usage by Sex')
plt.xlabel('Drug')
plt.ylabel('Count')
plt.legend(title='Sex', loc='upper right')
plt.xticks(rotation=45)
plt.show()
```
